```python
### 데이터 처리(조작) 라이브러리
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

import seaborn as sns

plt.rc("font", family="Malgun Gothic")

plt.rcParams["axes.unicode_minus"] = False
```

```python
file_path = "./files_org/SAMPLE_DB_20TABLE_2013.txt"
### sep="," : 데이터 구분자 정의 (기본값은 콤마, 생략가능)
files_org = pd.read_csv(file_path, sep=",")
files_org.head(1)
```

```python
file_path = "./files_org/표본자료 변수 테이블(2012-2013).xlsx"
files_trans_org = pd.read_excel(file_path, usecols="A,D")
files_trans_org

files_org_dict = dict(files_trans_org.to_numpy())

files_org_kr = files_org.rename(columns=files_org_dict)
files_org_kr.head(3)
```

```python
files_org_kr.info()
```

```python
files_org_kr.isnull().sum(axis=0)
```

```python
files_org_kr["최초입원일자"] = files_org_kr["최초입원일자"].fillna(0)
files_org_kr["요양만료일자"] = files_org_kr["요양만료일자"].fillna(0)
```

```python
files_org_kr.isnull().sum(axis=0)
```

```python
# '요양개시일자' 컬럼을 날짜 형식으로 변환 (연도와 월 추출을 위해)
files_org_kr['요양개시일자'] = pd.to_datetime(files_org_kr['요양개시일자'], format='%Y%m%d')
```

```python
# '연도'와 '월'을 추출하여 새로운 컬럼 추가
files_org_kr['연도'] = files_org_kr['요양개시일자'].dt.year
files_org_kr['월'] = files_org_kr['요양개시일자'].dt.month
```

```python
file_path2 = "./files_org/표본자료 변수 테이블(2012-2013).xlsx"
f_ref = pd.read_excel(file_path2, sheet_name='표2')
f_ref
```

## 이름바꾸기 샘플링

```python
files_org_kr['공상구분코드'].unique()
```

```python
f_ref.head()
```

```python
# 두 데이터프레임의 공상구분코드 값들을 비교
print(files_org_kr['공상구분코드'].unique())
print(f_ref['공상구분코드'].unique())
```

```python
# 데이터 타입 확인
print(files_org_kr['공상구분코드'].dtype)
print(f_ref['공상구분코드'].dtype)

# 만약 다르면 문자열로 변환
files_org_kr['공상구분코드'] = files_org_kr['공상구분코드'].astype(str)
f_ref['공상구분코드'] = f_ref['공상구분코드'].astype(str)
```

```python
f_kr_merge = pd.merge(left=files_org_kr,
                      right=f_ref,
                      how="left",
                      left_on="공상구분코드",
                      right_on="공상구분코드")
f_kr_merge
```

```python
# 공상구분명에 대한 비율 계산
vc = f_kr_merge['공상구분명'].value_counts(normalize=True)
```

```python
vc
```

```python
# seaborn을 사용하여 barplot 그리기
plt.figure(figsize=(21, 9))
plt.bar(vc.index, vc.values, color=sns.color_palette('pastel', len(vc)), width=0.6)

# 제목과 축 레이블 설정
plt.title('공상구분명 비율', fontsize=20)
plt.xlabel('공상구분명', fontsize=17)
plt.ylabel('비율', fontsize=17)

plt.grid(True, linestyle='--', alpha=0.7)

# 그래프 회전 (가독성을 위해)
plt.xticks(rotation=20)
plt.tight_layout()

# 차트 보여주기
plt.show()

"""
<해석>
 1. 대부분의 데이터는 정상건에 집중
  - 대부분의 환자 보험처리는 정산건으로 접수되어 있다.
 2. 일부 다른건도 존재
  - 일부 다른건(희귀난치성질환 지원대상자, 차상위 만성질환:18세미만 본인부담경감대상자 등)도 존재합니다.
  - 일부 다른건 중에서는 희귀난치성질환 지원대상자의 비율이 높아보입니다.
"""
```
