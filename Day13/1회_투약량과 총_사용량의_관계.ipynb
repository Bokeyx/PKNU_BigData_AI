## 진료내역에서 1회 투약량과 총 사용량의 관계

```python
### 데이터 처리(조작) 라이브러리
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

import seaborn as sns

plt.rc("font", family="Malgun Gothic")

plt.rcParams["axes.unicode_minus"] = False
```

```python
file_path = './files_org/SAMPLE_DB_30TABLE_2013.txt'
### sep="," : 데이터 구분자 정의 (기본값은 콤마, 생략가능)
files_org = pd.read_csv(file_path, sep=",")
files_org.head(1)
```

```python
files_org.info()
files_org.describe()
```

```python
file_path = './files_org/표본자료 변수 테이블(2012-2013).xlsx'
files_trans_org = pd.read_excel(file_path, usecols='A,D', sheet_name='30table_진료내역')
files_trans_org

files_trans_org.info()
files_trans_org.head(1)
```

```python
files_org_dict = dict(files_trans_org.to_numpy())
```

```python
files_org_kr = files_org.rename(columns=files_org_dict)
files_org_kr.head(3)
```

```python
files_org_kr.info()
```

```python
dose_per_day = np.array(files_org_kr['1일투약량']) # 1일 투약량
total_dose = np.array(files_org_kr['총사용량또는실시횟수']) # 총 사용량
```

```python
# 산점도 (Scatter plot) 그리기
plt.figure(figsize=(8, 6))
plt.scatter(files_org_kr['1회투약량'], files_org_kr['총사용량또는실시횟수'], color='blue', alpha=.25)

# 그래프 제목 및 축 레이블 설정
plt.title('진료내역에서 1회 투약량과 총 사용량의_관계', fontsize=14)
plt.xlabel('1회 투약량', fontsize=12)
plt.ylabel('총 사용량 또는 실시 횟수', fontsize=12)

# 그래프 출력
plt.grid(linestyle="--")
plt.show()

"""
<해석>
    1. 대부분의 데이터가 좌측 하단(낮은 투약량 & 낮은 총 사용량)에 집중
    	- 대부분의 환자는 1회 투약량이 적고, 총 사용량도 적은 경향을 보입니다.
	2. 일부 이상치(outlier) 존재
    	- X축(1회 투약량)이 500 이상인 데이터가 극소수 존재합니다.
    	- Y축(총 사용량)이 2500 이상인 경우도 보입니다.
    	- 특정한 경우에만 매우 높은 투약량이 사용되었음을 의미합니다.
	3. 1회 투약량과 총 사용량 간의 양의 상관관계
    	- 1회 투약량이 많아질수록 총 사용량도 증가하는 경향을 보입니다.
    	- 하지만 일부 데이터는 1회 투약량이 적어도 총 사용량이 많으며, 이는 장기간 소량씩 복용한 경우로 해석될 수 있습니다.
    4. 일부 이상치 지속 관찰이 필요할수도 있습니다.
"""
```
