```python
### 텐서플로우 프레임워크(패키지라고도 칭함)
import tensorflow as tf

### 케라스 라이브러리 불러들이기
from tensorflow import keras

### 시각화 라이브러리 불러들이기
import matplotlib.pyplot as plt

### 넘파이
import numpy as np

### 훈련 및 테스트 데이터로 분류하는 라이브러리
from sklearn.model_selection import train_test_split


### 딥러닝 랜덤 규칙 정의하기
# - 딥러닝에서의 랜덤 규칙은 항상 일정하게 유지되지는 않음
# - 딥러닝 모델 내부에서 훈련을 위한 데이터를 임의로 추출하여 사용(사람이 관여 안함)
tf.keras.utils.set_random_seed(42)
```

```python
(train_input, train_target), (test_input, test_target) = keras.datasets.fashion_mnist.load_data()

print(train_input.shape, train_target.shape)
print(test_input.shape, test_target.shape)
```

```python
train_scaled_255 = train_input / 255.0
test_scaled_255 = test_input / 255.0

print(train_scaled_255.shape, train_target.shape)
print(test_scaled_255.shape, test_target.shape)
```

```python
train_scaled, val_scaled, train_target, val_target = train_test_split(train_scaled_255, train_target, test_size=0.2, random_state=42)

print(train_scaled.shape,   train_target.shape)
print(test_scaled_255.shape, test_target.shape)
print(val_scaled.shape,     val_target.shape)
```

```python
model = keras.Sequential([
    keras.layers.Flatten(),
    keras.Input(shape=(28,28)),
    keras.layers.Dense(units=100, activation="sigmoid", name="Input_Layer"),
    keras.layers.Dense(units=70, activation="tanh", name="Hidden1_Layer"),
    keras.layers.Dense(units=50, activation="relu", name="Hidden2_Layer"),
    keras.layers.Dense(units=10, activation="softmax", name="Output_Layer")
])

adam = keras.optimizers.Adam(
    learning_rate= 0.001
)

model.compile(
    optimizer = adam,
    loss = "sparse_categorical_crossentropy",
    metrics = "accuracy"
)

model.fit(train_scaled, train_target, epochs=100, batch_size=128)
```

```python
train_score = model.evaluate(train_scaled, train_target)
val_score = model.evaluate(val_scaled, val_target)

print(f"훈련 손실율 : {train_score[0]}, 훈련 정확도 : {train_score[1]}")
print(f"검증 손실율 : {val_score[0]}, 검증 정확도 : {val_score[1]}")
```

```python
def optimizer_practice () :
    
    optimizer_classes = {
        "SGD": keras.optimizers.SGD,
        "Adagrad": keras.optimizers.Adagrad,
        "RMSprop": keras.optimizers.RMSprop,
        "Adam": keras.optimizers.Adam
    }
    
    activation_classes = {
        "sigmoid", 
        "tanh",
        "relu"
    }
    
    units_rates = [50, 60, 70, 80, 100]

    learning_rates = [0.001, 0.01]
    
    results = {}
    
    for opt_name, opt_class in optimizer_classes.items():
        for i in learning_rates:
            print("\n------------------------------------------------------------\n")
            print(f">>> Optimizer: {opt_name}, Learning Rate: {i}")
            
            if opt_name == "SGD":
                optimizer = opt_class(learning_rate=i, momentum=0.9)
            else:
                optimizer = opt_class(learning_rate=i)
                
            for act_name in activation_classes:
                for y in units_rates:
                    model = keras.Sequential([
                        keras.Input(shape=(28, 28)),
                        keras.layers.Flatten(),
                        keras.layers.Dense(units=y, activation=act_name),
                        keras.layers.Dense(10, activation="softmax")
                    ])
                    
                    model.compile(
                        optimizer=optimizer,
                        loss="sparse_categorical_crossentropy",
                        metrics=["accuracy"]
                    )

                    history = model.fit(train_scaled, train_target, epochs=10, batch_size=512)
                    
                    train_score = model.evaluate(train_scaled, train_target)
                    val_score = model.evaluate(val_scaled, val_target)
                    
                    print(f"훈련 손실율 : {train_score[0]}, 훈련 정확도 : {train_score[1]}")
                    print(f"검증 손실율 : {val_score[0]}, 검증 정확도 : {val_score[1]}")
                    

                    results[f"{opt_name}{act_name}{i}"] = val_score[0]
                    
                    best = min(results, key=results.get)
                    worst = max(results, key=results.get)

                    print(f"손실율이 가장 낮은 옵티마이저: {best} (손실: {results[best]:.2f})")
                    print(f"손실율이 가장 높은 옵티마이저: {worst} (손실: {results[worst]:.2f})")

    return results
    
```
