### Data Preprocssing for Classification Analysis
##### Combine by Data character

```python
"""
<Combine data each Length and Weight>
 - Combine bream and smelt Length Data
 - Combine bream and smelt Weight Data
"""
```

```python
"""
Make the data type for Training(Learning)
 - Problem Data: [[length, weight], [length, weight]....]
  -> The answer data is referred to as [independent variable, x].
  -> The independent variable (x) must be in a 2-dimensional format. This is a requirement for machine learning.
 
 - Answer Data: [bream, bream,...., smelt, smelt....]
  -> The answer data is referred to as [dependent variable, y].
  -> The dependent variable (y) must be in a 1-dimensional format. This is a requirement for both machine learning and deep learning.
  -> The dependent variable is used as the classification criterion.
   -> When classifying between two options, it is referred to as "binary classification".
      (Binary classification involves classifying between 0 or 1.)
   -> When classifying among three or more options, it is referred to as "multi-class classification".
      (Multi-class classification involves classifying between 0, 1, 2, ... n.)

* All feature data used for training (learning) must be in numerical form.
 - if categorical data is in text format, it should be converted into numbers (ranging from 0 to n).
  -> One-hot encoding can also be used for this conversion.
  -> Generally, a custom code is written to directly convert text data into numbers.

* Analysis Term
 - Each data (e.g., bream, smelt) is referred to as a "feature."
   (Feature, column, item, or future in deep learning all have the same meaning.)
"""
```

```python
### Combine bream and smelt Length Data
# - Variable name : Length
length = bream_length + smelt_length
len(length), print(length)
```

```python
### Combine bream and smelt Weight Data
# - Variable name : Length
weight = bream_weight + smelt_weight
len(weight), print(weight)
```

### Independent variable Preprocessing

```python
### Create 2 demensional independent variable processing by using length and weight data
# - [[length, weight], [length, weight]....]: 49row, 2column
# - variable name : fish_data
```

```python
### In Basic way to do
# Clean Variable list
fish_data = []
for idx in range(0, len(length), 1) :
    fish_data.append([length[idx], weight[idx]])

print(fish_data)

# Different Method 1
fish_data = []
for l, w in zip(length, weight) :
    fish_data.append([l, w])

# Different Method 2
fish_data = [[l, w] for l, w in zip(length, weight)]
len(fish_data), print(fish_data)
```

```python
### Create Dependent Variable
# - If there is a dependent variable in the collected data other than the independent variables, use it as it is.
# - If there isn't, you need to create one (but only if the dependent variable values are known...).
# - You must create the dependent variable values corresponding to the index positions of the independent variables.
# - Make it one-dimensional.
# - Use 1 for "bream" and 0 for "smelt" as data values.
# - Variable name: fish_target

fish_target = [1] * len(bream_length) + [0] * len(smelt_length)
print(len(fish_target), (fish_target))
```

```python
### Independent and Depndent have been defined for using on Training(learning)
# - Independent variable(x, length, weight): fish_data
# - Dependent variable(y, answer, type of fish): fish_target
# ** title of analysis: Identify Fish type by using weight and length of fish
```

### Set Learning

```python
"""
<Learning Model for Use in Learning>
 - Model Library: KNeighborsClassifier
 
 - Model Package: sklearn.neighbors
 
 - Algorithm Name: Call it "KNN"
  -> K-Nearest Neighbors (This is a model used for the KNeighbors algorithm)
  
 - Among machine learning models, KNN is the simplest one.
 - It is mostly used for comparison with other models.

<Process Overview>
 - Classification (judging, predicting) by comparing the feature values of interest with the closest neighbors' values and distances.
 - The number of nearest neighbors (the number of neighbors) is used as a basis for comparison against the ratio (probability).
 - The prediction is made based on the highest ratio (majority vote principle).
 
<Machine Learning Learning Sequence>
 1. Collect Data
 2. Data Preprocessing and inquiry
 3. Data Transformation
  - Create independent variables and dependent variables
  - Divide into training data and test data (or divide into training, validation, and test data)
 4. Create Training Model
 5. Train the Model
  - Using learning data
 6. Validate Model Accuracy
   - Using Test Data (or Validation Data)
   - Check the accuracy before training and compare it with the accuracy of the test (or validation) data after training.
  6-1. Tuning
  6-2. Performance Evaluation
 7. Make Predictions (Verify how well the model predicts with new data)
   - Random data, i.e., test data.
  7-1. Final accuracy verification (Test accuracy verification).
  7-2. Performance evaluation.
"""
```

```python
### Model to use
from sklearn.neighbors import KNeighborsClassifier
```

```python
### 1. Create Training Model (it is class)
kn = KNeighborsClassifier()
kn
```

```python
### 2. Train the model
# - Supervised learning training function: fit(independent variable, dependent variable)
# - Unsupervised learning training function: fit(independent variable)
# - The model that is trained is referred to as the "trained model."
# - The more data there is, the longer the training time will be (depends on PC performance).
# - The independent and dependent variables used in fit are referred to as "training data."
kn.fit(fish_data, fish_target)
```

```python
### Validate Model Accuracy
# - Check accuracy with training data and random data (validation or test data).
# - Among the data used in score(),
#  -> When the training data is used to check accuracy, it is referred to as training accuracy.
#  -> When validation or test data is used to check accuracy, it is referred to as validation accuracy or test accuracy.
# - Score() result: The result ranges from 0 to 1 (0% to 100%). The closer to 1, the higher the accuracy.
# (analysis) If it shows training accuracy of 100%, it indicates that the model has been trained very well.
#  - In practice, if the training accuracy is 100%, the model is considered unusable.
#   -> Probabilities up to 99.999% are accepted, but 100% is not considered acceptable.
kn.score(fish_data, fish_target)
```

```python
### Make Predictions
# - Temporary Data using for predictions: length 30cm, weight 600g
```

```python
### Use a scatter plot to check the location of the predicted data.
# - Draw separate scatter plots for bream and smelt (as shown above).
# - Additionally, display the predicted data as a scatter plot (with green color).
```

```python
###  Make scatter plot by using bream and smelt data in one graph
# - Color: bream = red, smelt = blue
# - Show legend
plt.title("Length & Weight of Bream vs. smelt")
plt.scatter(bream_length, bream_weight, color="red", label="Bream")
plt.scatter(smelt_length, smelt_weight, color="blue", label="Smelt")
plt.scatter(x=30, y=600, color="green", marker="^")
plt.xlabel("Length")
plt.ylabel("Weight")
plt.grid(linestyle = "--")
plt.legend()
plt.show()
```

```python
### Use Predict Function
# - predict Function: predict([[Independent variable for prediction]])
# - The input must be provided as a 2-dimensional array (same dimensions as the independent variables).
kn.predict([[30, 600]])
# 1 represents "bream", and 0 represents "smelt" — following the rules (category values) used when creating the dependent variable.
```

### Use Hyperparameter

```python
"""
<Meaning of Hyperparameter>
 - This refers to manual tuning where a person adjusts specific parameter values to improve the model's performance.
 - The parameters that affect the performance (accuracy) of the trained model are called hyperparameters.
 - By changing (tuning) the values of hyperparameters, you can improve the model's performance.

<Hyperparameter Tuning>
 - This refers to hyperparameter tuning, where the values of parameters are adjusted and training-validation 
   is performed to improve the model's performance.

<Hyperparameters used in KNN>
 - n_neighbors: The number of neighbors can be changed to improve performance (default value = 5).
 - This is specified as an attribute when creating the model.
"""
```

```python
### Create Training model
# - Number of neighbors: Set to 20.
kn20 = KNeighborsClassifier(n_neighbors=20)
kn20
```

```python
### 2. Train the model
kn20.fit(fish_data, fish_target)
```

```python
### Validate Model Accuracy
# - score() function process sequence
#  - It internally performs predict().
#  - Then, it compares the predict() results with the fish_target to check and display the accuracy.
train_score = kn20.score(fish_data, fish_target)
round(train_score, 3)

### In analysis documentation, the number of decimal places used is expressed up to 3 decimal places.
```

### Tuning Hyperparameter

```python
"""
When the training accuracy is not 1 but the highest accuracy is achieved, 
the number of neighbors can be identified. This allows you to select the best hyperparameter value.
"""
```

```python
### Create model with the number of neighbors set to the default value.
kn = KNeighborsClassifier()
kn
```

```python
### Train the model
kn.fit(fish_data, fish_target)
```

```python
train_score = kn.score(fish_data, fish_target)
train_score
```

```python
### Tuning the hyperparameter value of the number of neighbors.
# - The number of neighbors should be odd to ensure a majority vote can be formed.
kn.n_neighbors = 20
train_score = kn.score(fish_data, fish_target)
round(train_score, 3)
```

```python
### Find the number of neighbors when the accuracy is the highest but not equal to 1...

# Variable to store the number of neighbors when the accuracy is the highest but not 1
nCnt = 0
# Variable to store the highest accuracy value when it's not 1
nScore = 0.0

for n in range(3, len(fish_data), 2):
    kn.n_neighbors = n
    train_score = kn.score(fish_data, fish_target)

    ### Condition check
    # When the accuracy is not 1
    if train_score < 1:
        if nScore < train_score:
            nScore = train_score
            nCnt = n

print(round(nScore, 3), nCnt)
```

```python
kn = KNeighborsClassifier()

# Variable to store the best accuracy
best_score = 0  
best_n_neighbors = 0 

# The value of n_neighbors should be odd, so only test odd values
for i in range(3, len(fish_data) + 1, 2): 
    kn.n_neighbors = i 
    kn.fit(fish_data, fish_target)
    train_score = kn.score(fish_data, fish_target)
    
    # Find the largest value where accuracy is less than 1
    if train_score < 1 and train_score > best_score:
        best_score = train_score
        best_n_neighbors = i

print(round(best_score, 3), best_n_neighbors)
```
