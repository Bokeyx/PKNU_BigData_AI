```python
### 데이터 처리(조작) 라이브러리
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

import seaborn as sns

plt.rc("font", family="Malgun Gothic")

plt.rcParams["axes.unicode_minus"] = False
```

```python
### 엔스콤() 데이터 시각화
# - seaborn 라이브러리 내에 ansvombe 데이터셋을 제공하고 있음
ans = sns.load_dataset("anscombe")
ans
```

```python
ans.info
ans.describe()
```

## dataset 컬럼의 범주형 데이터를 그룹화(집계) 하기

```python
### groupby 함수 사용
# 그룹 변수명 : ans_group
# - 범주형 컬럼인 dataset 컬럼을 기준으로 그룹만 생성합니다.
#  (별도의 집계함수를 하용하지 않아도 됩니다.)
ans_group = ans.groupby(["dataset"])
ans_group
```

```python
### 그룹(집계)된 데이터의 확인은 had() 함수를 통해서 확인합니다.
# - head(번호) : head안에 번호는 각 그룹 내에서 조회할 행의 갯수를 의미
# - 그룹별로 행의 갯수 만큼 조회해 줍니다
ans_group.head(2)
```

```python
### 그룹별로 기초통계 확인하기
# - 그룹화된 데이터는 각 컬럼별 그룹화되어 기초통꼐를 출력해 줍니다.
ans_group.describe()
```

## matplolib 라이브러리에서 제공하는 시각화

```python
### 선그래프
plt.plot(tips["total_bill"])
# plt.show()

plt.plot(tips["tip"])
plt.show()
```

```python
# 그래프 크기 설정
plt.figure(figsize=(12, 8))

# 바 그래프 그리기 (에러바 제거)
bars = sns.barplot(x="sex", y="total_bill", data=tips, errorbar=None)

# 각 성별 그룹에 대해 텍스트 추가하기
for bar in bars.patches:  # `bars.patches`를 통해 각 바의 높이를 접근
    height = bar.get_height()  # 바의 높이 가져오기
    x_pos = bar.get_x() + bar.get_width() / 2  # 바의 x 위치 중앙으로 설정
    # 텍스트를 바 위에 자연스럽게 배치
    plt.text(
        x=x_pos,  # 바 중앙의 x 위치
        y=height + 0.5,  # 바의 높이 바로 위에 텍스트 위치
        s=str(round(height, 2)),  # 바의 높이 값을 텍스트로 표시
        color="black",  # 텍스트 색상
        ha="center",  # 수평 정렬
        va="bottom"  # 수직 정렬 (바 위에 텍스트가 표시됨)
    )

# 그래프 제목 및 설정
plt.title("성별 지출액 빈도 분석 그래프")
plt.grid(linestyle="--")

# 그래프 표시
plt.show()
```

```python
### 그래프 여러개 동시에 그리기
# jointplot() 함수 사용
# - 함수 내부에서 속성으로 그래프 형태 정의
# - 메인 그래프는 삼점도로, 상단 및 우측에는 히스토그램 막대그래프가 나타남
sns.jointplot(x="total_bill", y="tip", data=tips, kind="scatter")
plt.show
```
