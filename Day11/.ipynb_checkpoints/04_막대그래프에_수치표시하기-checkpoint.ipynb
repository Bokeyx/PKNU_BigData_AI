```python
### 데이터 처리(조작) 라이브러리
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

import seaborn as sns

plt.rc("font", family="Malgun Gothic")

plt.rcParams["axes.unicode_minus"] = False
```

```python
### 샘플데이터
x = [1, 2, 3, 4, 5]
y = [100, 200, 300, 400, 500]

### bar(막대) 그래프 그리기
# - ec(edgecolor) : 막대그래프 테두리 색상 지정
# - alpha : 막대그래프 내부 및 테두리 투명도 (0~1사이의 값 사용)
# -막대그래프 각각의 color 지정
plt.bar(x, y, ec="black", alpha=0.5, color=["red", "blue", "orange", "green", "yellow"])

### 각 막대별로 x축 및 y축의 위치에 value값 표시하기
# - enumerate : 리스트의 인덱스와 각 인덱스의 값을 튜플로 셋트를 만들어주는 함수
#             : [(인덱스번호), value 값, ()....]
#             : [(0, 1), (1, 2), (2, 3), (3, 4), (4, 5)]
#             : zip() 함수와 동일한 결과 형태로 반환해 줍니다.
for idx, x_v in enumerate(x):
    # print(f"{idx}, {x_v}")

    ### 각 막대에 y축의 값 추가하기
    plt.text(
        ### 텍스트가 위치할 x축 위치 값 정의(x 좌표)
        x_v,
        ### 텍스트가 위치할 y축 위치 값 정의(y 좌표)
        y[idx],
        ### 텍스트로 표시할 값 정의
        y[idx],
        ### 텍스트 color(색상) 정의
        color = "blue",
        ### 텍스트 수평정령 하기
        # - ha(horizontalalignment)
        # - left, center, right
        ha="center",
        ### 텍스트 수직정령 하기
        # - va(verticalalignment)
        # - top, center, botton
        va="bottom"
    )

plt.show()
    
```

```python
### 위 그래프 예제를 그래도 사용
# - 각 막대의 텍스트의 색상(color)dmf 막대의 색상(color와 동일하게 처리해 주세요.


### 샘플데이터
x = [1, 2, 3, 4, 5]
y = [100, 200, 300, 400, 500]

### bar(막대) 그래프 그리기
# - ec(edgecolor) : 막대그래프 테두리 색상 지정
# - alpha : 막대그래프 내부 및 테두리 투명도 (0~1사이의 값 사용)
# -막대그래프 각각의 color 지정
plt.bar(x, y, ec="black", alpha=0.5, color=["red", "blue", "orange", "green", "yellow"])

### 각 막대별로 x축 및 y축의 위치에 value값 표시하기
# - enumerate : 리스트의 인덱스와 각 인덱스의 값을 튜플로 셋트를 만들어주는 함수
#             : [(인덱스번호), value 값, ()....]
#             : [(0, 1), (1, 2), (2, 3), (3, 4), (4, 5)]
#             : zip() 함수와 동일한 결과 형태로 반환해 줍니다.

c2=["red", "blue", "orange", "green", "yellow"]
for idx, x_v in enumerate(x):
    # print(f"{idx}, {x_v}")

    ### 각 막대에 y축의 값 추가하기
    plt.text(
        ### 텍스트가 위치할 x축 위치 값 정의(x 좌표)
        x_v,
        ### 텍스트가 위치할 y축 위치 값 정의(y 좌표)
        y[idx],
        ### 텍스트로 표시할 값 정의
        y[idx],
        ### 텍스트 color(색상) 정의
        color = c2[idx],
        ### 텍스트 수평정령 하기
        # - ha(horizontalalignment)
        # - left, center, right
        ha="center",
        ### 텍스트 수직정령 하기
        # - va(verticalalignment)
        # - top, center, botton
        va="bottom"
    )

plt.show()
    
```
