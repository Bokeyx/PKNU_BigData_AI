## 국적 컬럼의 데이터에서 대륙 처리하기

```python
### 국적 컬럼의 데이터에서 "고유한 값"들만 조회하기
kto_201901["국적"].unique()

### 대륙이름
# - '아시아주', '미주', '구주', '대양주', '아프리카주', '기타대륙', '교포소계'
# - 컬럼 생성 시 사용 :
#   '아시아', '아메리카', '유럽', '오세아니아', 아프리카', '기타대륙', '교포'
#      25         5         23         3           2           1         1 
```

```python
### 실제 국적만 추출하기
### 국적 데이터 중에 대륙이름이 포함되지 않는 데이터만 필터링 하기
c = kto_201901["국적"].isin(['아시아주', '미주', '구주', '대양주', 
                             '아프리카주', '기타대륙', '교포소계'])
kto_201901_country = kto_201901[c == False]
kto_201901_country
```

```python
### 눈에 보이는 인덱스 값을 순차적으로 재정의 하기
# - 주로 행을 삭제하거나 ,추가하는 경우에 인덱스 값 재정의를 수행합니다.
# - drop=True : 기존 인덱스 번호를 컬럼으로 사용하지 않겠다는 의미
kto_201901_country_newindex = kto_201901_country.reset_index(drop=True)
kto_201901_country_newindex
```

## 대륙 컬럼 추가하기

```python
### 대륙이름
# - '아시아주', '미주', '구주', '대양주', '아프리카주', '기타대륙', '교포소계'
# - 컬럼 생성 시 사용 :
#   '아시아', '아메리카', '유럽', '오세아니아', '아프리카', '기타대륙', '교포'
#      25         5         23         3           2           1         1 
```

```python
### 총 국적 데이터 갯수 60개 국가
len(kto_201901_country_newindex)
```

```python
### 컬럼에 추가 할 대륙이름을 리스트로 생성하기
continents = ['아시아']*25 + ['아메리카']*5 + ['유럽']*23 + ['오세아니아']*3 + \
             ['아프리카']*2 + ['기타대륙'] + ['교포']
print(continents, len(continents))
```

```python
### "대륙" 컬럼에 위 리스트 데이터 추가하기
kto_201901_country_newindex["대륙"] = continents
kto_201901_country_newindex
```

## 관광객비율(%) 컬럼 추가하기

```python
### 각 국가내에서 관광을 위해 입국하는 비율(%)을 계산하여 컬럼 데이터로 사용
kto_201901_country_newindex.head(2)
```

```python
### 국가별로 -> 관광을 위해 입국하는 비율 계산
kto_201901_country_newindex["관광객비율(%)"] = \
    round(kto_201901_country_newindex["관광"] / kto_201901_country_newindex["계"] * 100, 3)
```

## 전체관광객비율(%) 컬럼추가하기

```python
### 모든 국가의 전체 광객수를 기준으로 각 국가의 관광객수로 나누어서 비율 계산
# 관광 / sum(관광)
kto_201901_country_newindex.head(3)
```

```python
### 관광 컬럼의 모든 데이터 sum하기
t_sum = kto_201901_country_newindex["관광"].sum()
t_sum
```

```python
kto_201901_country_newindex["전체관광객비율(%)"] = \
    round(kto_201901_country_newindex["관광"] / t_sum * 100, 3)

kto_201901_country_newindex.head(3)
```

## 116개 파일 -> 행단위 통합하기

```python
### 최종 통합에 사용 할 Dataframe명 : df
#    - 깨끗한 데이터프레임 정의 : df.DataFrame()
#    - 누적 추가 : df = pd.concat([,], ...)
### 통합 후 파일 저장명 : : kto_total.xlsx
```

```python
### 깨끗한 데이터프레임 정의(통합본)
### 116개의 파일을 읽어 들일 수 있도록 파일명 생성하기
# - 파일명의 패턴을 반복문을 이용해서 사용(년도, 월)
```

```python
df = pd.DataFrame()

for y in range(2010, 2020) :
    for m in range(1, 13) :
        file_path = f"./files2/kto_{y}{str(m).zfill(2)}.xlsx"
### 예외처리 try~except
# - 외부 영역에서 어떤 오류가 발생할지 모르는 경우에 예외처리 합니다.
#print(file_path)
        try :
            kto_201901 = pd.read_excel(file_path,
                                       header=1,
                                       skipfooter=4,
                                       usecols="A:G")
            
            # 2. [기준년월] 컬럼 추가
            kto_201901["기준년월"] = f"{y}-{str(m).zfill(2)}"
            
            # 3. [국적] 컬럼 추가 [대륙] 이름 제거
            c = kto_201901["국적"].isin(['아시아주', '미주', '구주', '대양주', 
                             '아프리카주', '기타대륙', '교포소계'])
            kto_201901_country = kto_201901[c == False]
            
            # 4. "인덱스" 재정의
            kto_201901_country_newindex = kto_201901_country.reset_index(drop=True)
            
            # 5. [대륙] 컬럼 추가
            continents = ['아시아']*25 + ['아메리카']*5 + ['유럽']*23 + ['오세아니아']*3 + \
            ['아프리카']*2 + ['기타대륙'] + ['교포']

            kto_201901_country_newindex["대륙"] = continents
            
            # 6. [관광객비율(%)] 컬럼 추가
            kto_201901_country_newindex["관광객비율(%)"] = \
            round(kto_201901_country_newindex["관광"] / \
                  kto_201901_country_newindex["계"] * 100, 3)

            
            # 7. [전체관광객비율(%)] 컬럼 추가
            t_sum = kto_201901_country_newindex["관광"].sum()
            kto_201901_country_newindex["전체관광객비율(%)"] = \
            round(kto_201901_country_newindex["관광"] / t_sum * 100, 3)

            ## 8. 원본 df 데이터프레임에 위 데이터프레임 행단위 추가하기
            df = pd.concat([df, kto_201901_country_newindex], ignore_index=True)
            
        except :
            ### 오류가 발생할 경우 처리되는 영역
            break
```
