```python
import pandas as pd
```

```python
### bicycle_dup.csv 파일 읽어들이기
# 변수명 : df

file_path = "./data/bicycle_dup.csv"
df = pd.read_csv(file_path, encoding="euc-kr")

df.info()
df.describe()
```

## 결측치 처리

```python
### 결측치가 있는 컬럼들은 대체하기
# - 각 컬럼의 [전체 평균값]으로 대체하기
# - 결측치가 있는 컬럼 : 대여거치대, 반납거치대, 이용시간, 이용거리
```

```python
### 각각 처리 코드
df["대여거치대"].fillna(df["대여거치대"].mean())
df["반납거치대"].fillna(df["반납거치대"].mean())
df["이용시간"].fillna(df["대여거치대"].mean())
df["이용거리"].fillna(df["이용거리"].mean())

### 응용처리
cols = ["대여거치대", "반납거치대", "이용시간", "이용거리"]
for col in cols :
    df[col].fillna(df[col].mean())

### 간단처리
df.fillna(df[["대여거치대", "반납거치대", "이용시간", "이용거리"]].mean())
```

```python
df = df.fillna(df[["대여거치대", "반납거치대", "이용시간", "이용거리"]].mean())
df.info()
```

## 전체 행단위 중복 데이터 확인하기

```python
### duplicated() : 중복 체크 함수
# - 중복이 있는 부분은 True, 없는 부분은 False
df.duplicated()

### 중복이 있는 True의 총갯수와 중복이 없는 False의 총갯수 확인하기
df.duplicated().value_counts()
```

## 특정 컬럼을 기준으로 행단위 중복확인

```python
### 이용거리 컬럼을 기준으로 중복 확인
# - 중복 데이터 중에 첫번째 중복값은 False,
#                  두번째 이상 중복값들은 모두 True가 됩니다.
# - 중복값 중 에 하나는 존재해야하기 때문에...
df["이용거리"]
```

```python
df["이용거리"].duplicated()
df.duplicated(["이용거리"])
```

```python
### 이용거리 컬럼을 기준으로 중복 확인
# - 이용거리 컬럼에서 중복이 없는, 중복이 없는 모든 컬럼의 데이터 조회해 주세요
```

```python
c = df.duplicated(["이용거리"])
df[c == False]
```

## 중복값 선택하기

```python
### keep : 중복에서 제외할 위치 지정
# - first : 첫번째 중복값은 중복에서 제외(첫번째 이후 부터는 중복 처리)
#         : 기본값은 생략가능
# - last : 마지막 중복값은 중복에서 제외(마지막 이전 데이터는 모두 중복 처리)
# - False : 모두 중복 처리
df.duplicated(["이용거리"], keep = "first")
```

```python
df.duplicated(["이용거리"], keep = "last")
```

```python
df.duplicated(["이용거리"], keep = False)
```

## 중복데이터 처리하기

```python
"""
<중복데이터 처리 방법>
 1. 중복값 확인
  - 중복을 허용하는 데이터인지 확인 필요(데이터를 제공해 주시분들에게)
 2. 중복값을 삭제할지, 그냥 사용할지 결정
  * 삭제할 경우
   -> 중복된 행을 삭제합니다.
  * 그냥 사용할 경우
   -> 실제로 중복이 가능한 데이터의 성격인 경우가 해당됩니다.
"""
```

```python
### 중복 데이터 삭제하기
# - 행 단위 삭제가 진행됩니다.

### 행 전체 중복이 있는 경우, 해당 행 삭제
# - 단 keep의 중복 위치 조건에 따라 True인 행만 삭제
df.drop_duplicates(keep="first")
```

```python
### 이용거리 컬럼 데이터를 기준으로 중복된 행 삭제하기
# - 최초 중복은 중복에서 제외하기
df["이용거리"].drop_duplicates(keep="first")

### 중복 제거 후에 데이터를 다시 가지고 와야하기에, 아래 방법이 편한 방법이 될 수 있음
df.drop_duplicates(["이용거리"], keep="first")
```
