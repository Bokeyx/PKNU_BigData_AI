```python
### 5.Courses 컬럼의 0번째 행의 값 추출해주세요.
```

```python
df2_dict = df2["5.Courses"][0]
type(df2_dict)
```

```python
df2_dict
```

```python
"""
위 df2_dict 변수의 데이터를 이용해서 아래와 같이 출력해 주세요.
Classes : probability
Classes : Generalized Linear Model
Classes : Categorical Data Analysis
Major : Statistics
"""

### 딕셔너리 출력
for k, v in df2_dict.items() :
    if type(v) == list :
        for v2 in v:
           print(f"{k} : {v2}")
    else :
        print(f"{k} : {v}")
```

```python
df2_list = list(zip(df2_dict.keys(), df2_dict.values()))

for i in df2_list:
    if i[0] == 'Classes':
        for cn in i[1]:
            print(f"Classes : {cn}")
    else:
        print(f"{i[0]} : {i[1]}")
```
