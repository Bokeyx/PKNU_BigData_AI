```python
### Machine Running library install
from sklearn.neighbors import KNeighborsClassifier

### visualization library
import matplotlib.pyplot as plt

# Definition of the NumPy library
import numpy as np
```

```python
fish_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0, 
                31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0, 
                35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0, 9.8, 
                10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]

fish_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0, 
                500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0, 
                700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0, 6.7, 
                7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]

len(fish_length), len(fish_weight)
```

```python
### Independent Variable
fish_data = [[l, w] for l, w in zip(fish_length, fish_weight)]
len(fish_data)
fish_data
```

```python
### Dependent Variable
fish_target = [1]*35 + [0]*14
len(fish_target), print(fish_target)
```

```python
### Creating Independent Variables with the NumPy Library
# - column_stack: A function that stacks multiple lists by combining elements at the same index position into a list
# - Result type: NumPy array type
fish_data = np.column_stack((fish_length, fish_weight))
print(fish_data.shape)
print(fish_data)
```

```python
### Creating Dependent Variables Using NumPy Functions
# - concatenate(): Concatenate arrays -> Result is a NumPy array (array)
# - ones(35): Create 35 ones -> Returned as a array
# - zeros(14): Create 14 zeros -> Returned as a array
fish_target = np.concatenate((np.ones(35),np.zeros(14)))
print(fish_target.shape)
fish_target
```

```python
### Dataset to be used: Independent and Dependent Variables
fish_data.shape, fish_target.shape
```

### Splitting the Data While Shuffling (Using a Library)

```python
### Definition of Library (Preprocessing Library)
# - Functions used in data-based machine learning and deep learning: train_test_split()
# - A function that splits data while shuffling it randomly (Training:Testing) â€“ It splits the data into exactly two sets.
from sklearn.model_selection import train_test_split
```

```python
### Splitting Criteria -> Training : Testing = 75% : 25%
# - The splitting criterion is based on the "ratio."
# - Result value order: Returned in order of training independence, test independence, training dependency, and test dependency
#                                                 2 dimension,        2 dimension,         1 dimension,        1 dimension
train_input, test_input, train_target, test_target= train_test_split(
         # - Independent variable complete data
         fish_data, 
         # - Dependent variable complete data
         fish_target, 
         # - The ratio for splitting into training or testing (use either one)
         #train_size=0.75,
         test_size=0.25, 
         # - Random seed
         random_state=42,
         # - Adjust the ratio of the dependent variable's category values to prevent bias 
         #   (mainly used in classification analysis, not often in regression)
         stratify=fish_target
)
```

```python
### Check the data dimensions
print(train_input.shape, train_target.shape)
print(test_input.shape, test_target.shape)
```

### Training the model

```python
### Create Training Model
kn = KNeighborsClassifier()
kn
```

```python
### Training the model
kn.fit(train_input, train_target)
```

```python
### Check accuracy by Training Data : Training Accuracy
train_score = kn.score(train_input, train_target)
train_score
```

```python
### Check accuracy by Testing Data
test_score = kn.score(test_input, test_target)
test_score
```

```python
### Make predictions using the test data
test_pred = kn.predict(test_input)
print(test_pred)
print(test_target)
```

### Scaling processing (normalized == standardized)

```python
"""
<Scaling>
- It's called normalization or standardization
- Scaling processing time: Processes when there is a large unit deviation of data values between characteristics.
- In general, if you are conducting machine learning, you can think of it as an essential step.
  (First proceed without scaling > then proceed after scaling > compare the two training performances)
"""
```

```python
###Scaling Standardization Calculation Formula
# - Standardized = (Each data - data overall average) / data overall standard deviation
# - Standardized: A value that indicates how many standard deviations each data is from the origin (0)
# - Use a function of numpy for mean and standard deviation
```

```python
### Getting the overall mean of training independent variable data
# - Find and return the overall average for each of the two columns (length, weight) in the training independent variable data.
# - axis = 0: Means the average of the row units in each column
mean = np.mean(train_input, axis=0)
mean
```

```python
### Training Independent Variables Data Overall Standard Deviation
# - Returns the overall standard deviation for each length and weight
std = np.std(train_input, axis=0)
std
```

```python
### Standardized(Standard score) = (Each data - data overall average) / data overall standard deviation
train_scaled = (train_input - mean) / std
train_scaled.shape
```

```python
new = ([25, 150] - mean) / std
new[0], new[1]
```

```python
### Training independent variable standardized data and random data (25,150)
### Please express it with a scatter plot
### -- Changed to scaled data, so you don't have to think about the existing neighborhood

plt.title("Visualization of Training and Test Data Length and Weight")
plt.scatter(train_scaled[:, 0], train_scaled[:, 1], color="red", label="Training")
plt.scatter(new[0], new[1], color="green", marker="^", label="Prediction")
plt.xlabel("Length")
plt.ylabel("Weight")
plt.grid(linestyle = "--")
plt.legend()
plt.show()
```

```python
### Create Training Model
kn = KNeighborsClassifier()
kn

### Training the model
kn.fit(train_scaled, train_target)

### Check accuracy by Training Data
train_score = kn.score(train_scaled, train_target)
print(train_score)

### Check accuracy after scaling test independent variables
# - Test independent variables use the standard score created using the training independent variables
# - All data changes are made with training data, and other(verification or test) data are used only with training data
#   ***** Reason: We trained using standard score data made from training data, 
#                 so we need to return it to standard score made from training data with other data.
test_scaled = (test_input - mean) / std
test_score = kn.score(test_scaled, test_target)
print(test_score)

### Predicting with random data (25,150)
test_pred = kn.predict([new])
print(test_pred)
print(test_target)

### Extract used neighbor data
distances, indexes = kn.kneighbors([new])

print(f"Training Accuracy:{train_score}, Test Accuracy:{test_score}, used neighbor index:{indexes}")
```
