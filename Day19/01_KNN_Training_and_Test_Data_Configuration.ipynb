### Data Splitting: Training, Testing (or Training, Validation, Testing)

```python
"""
<Data Splitting>
 - Training Data: Used in fit() or score()
 - Validation Data: Used in score()
 - Test Data: Used in predict(), score()
 - If the data is split only into training and testing, use score() and predict() on the test data

<Reasons for Data Splitting>
 - To verify the model's performance and ensure proper training
 - To validate the reliability(credibility) of the validation process

<Splitting Criteria>
 - Split into 2 parts -> Training: Testing
 - Split into 3 parts -> Training: Validation: Testing

<Splitting Ratios>
 - Split into 2 parts -> Training(7): Testing(3) <- Most commonly used
                       -> Training(7.5): Testing(2.5)
                       -> Training(6): Testing(4)
 - Split into 3 parts -> Training(6): Validation(2): Testing(2) <- Most commonly used
                       -> Training(7): Validation(1.5): Testing(1.5)
                       -> Training(7): Validation(2): Testing(1)

<Commonly Used Variable Names>
 * Training Data (train)
   - Training Independent Variable Name: train_input, train_X, train_data
   - Training Dependent Variable Name: train_traget, train_y

 * Validation Data (train)
   - Validation Independent Variable Name: val_input, train_X, val_data
   - Validation Dependent Variable Name: val_traget, val_y

 * Testing Data (train)
   - Testing Independent Variable Name: test_input, train_X, test_data
   - TestingDependent Variable Name: test_target, test_y
"""
```

### Training: Separate into test data

```python
### Separate into 35 training data and 14 test data
# - Separate the first 35 as training data and the last 14 as test data for the independent and dependent variables
# Training variable name: train_input, train_traget
# Test variable name: test_input, test_traget

### Separate data into training and testing sets for independent and dependent variables
# Independent Data from Training Data
train_input = fish_data[:35]

# dependent Data from Training Data
train_target = fish_target[:35]

# Independent Data from Test Data
test_input = fish_data[35:]

# dependent Data from Test Data
test_target = fish_target[35:]

len(train_target), len(test_target), len(train_input), len(test_input)
```

### Training the model

```python
### Create KNN Training MOdel
# - Variable Name: kn
# - kn: The number of neighbors can be changed to improve performance (default value = 5).
kn = KNeighborsClassifier()
kn
```

```python
### Training the model
kn.fit(train_input, train_target)
```

```python
### Check accuracy by Training Data : Training Accuracy
# - The internal process of the score() function
#  -> It first processes predict(train_input) to get the predicted results -> [1, 1, ...]
#  -> It then compares the predicted results with the actual answers (train_target) and provides the accuracy.
train_score = kn.score(train_input, train_target)
train_score
```

```python
### Check accuracy by Testing Data
# - To verify if the model has trained successfully.
test_score = kn.score(test_input, test_target)
test_score
```

```python
### Make predictions using the test data
test_pred = kn.predict(test_input)
print(test_pred)
print(test_target)
```

### Randomly shuffle the data to resolve bias.

```python
# Definition of the NumPy library
import numpy as np
```

```python
### Shuffle the data using the original data.
# - Original Data -> Python List Type
len(fish_data), len(fish_target)
```

```python
### Convert the original data, which is of Python's list type, to a NumPy array type.
# - To shuffle randomly using NumPy: NumPy uses arrays, not lists.
input_arr = np.array(fish_data)
target_arr = np.array(fish_target)

### Check the number of data.
print(len(input_arr), len(target_arr))

### Check the dimensions of the data (only possible with NumPy array type).
# - (49, 2) = (49 row, 2 column), (49,) = (49 row, 1 column)
print(input_arr.shape, target_arr.shape)
```

```python
### Set the random rule.
# - Random library: random
# - Random rule function: seed()
# - Meaning of the number: Internally, NumPy generates random operations based on the seed number. 
#                        :The number itself has no particular meaning (e.g., the number 42 is commonly used, but it's just a convention).
# - It’s a good practice to set a random rule when using the random library.
#  -> When validating model performance multiple times or with different people, without a fixed random rule, 
#     the results may vary each time the code is run. To ensure consistent results, it’s recommended to apply a fixed seed.
#     (Even with a fixed rule, some values may still differ, especially in deep learning, where it's harder to achieve exact reproducibility.)
np.random.seed(42)
```

```python
### Create Random Index Numbers
# - To shuffle the input and target data while maintaining their mapping (alignment).
# - Generate index values from 0 to 48.
index = np.arange(49)
index
```

```python
### Shuffle the generated index values from 0 to 48.
np.random.shuffle(index)
index
```

```python
### Create the training target (35 items).

### Separate data into training and testing sets for independent and dependent variables
# Independent Data from Training Data
train_input = input_arr[index[:35]]

# dependent Data from Training Data
train_target = target_arr[index[:35]]

### Create the test target (14 items).

# Independent Data from Test Data
test_input = input_arr[index[35:]]

# dependent Data from Test Data
test_target = target_arr[index[35:]]

print(train_input.shape, train_target.shape, test_input.shape, test_target.shape)
```

```python
### checking if it shuffled well by making a scatter plot
# x-axis = length, y-axis = weight
# You can use the array() type in the same way as a list.
# Plot both the training and test data as scatter points on a single graph.
# Display training data in red and test data in blue.
plt.title("Visualization of Training and Test Data Length and Weight")
plt.scatter(train_input[:, 0], train_input[:, 1], color="red", label="Train")
plt.scatter(test_input[:, 0], test_input[:, 1], color="blue", label="Test")
# plt.scatter([i[0] for i in train_input], [i[1] for i in train_input], color="red", label="train")
# plt.scatter([t[0] for t in test_input], [t[1] for t in test_input], color="blue", label="test")
plt.xlabel("Length")
plt.ylabel("Weight")
plt.grid(linestyle = "--")
plt.legend()
plt.show()
```
