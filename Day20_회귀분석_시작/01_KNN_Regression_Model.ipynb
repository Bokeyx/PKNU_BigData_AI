```python
### Machine Running library install
from sklearn.neighbors import KNeighborsRegressor

### visualization library
import matplotlib.pyplot as plt

# Definition of the NumPy library
import numpy as np

### Definition of Library (Preprocessing Library)
from sklearn.model_selection import train_test_split
```

```python
"""
<Themes>
 - Using the length of the perch -> Estimate the weight

* Independent variable: length
* Dependent variable: weight

 - The data type of the dependent variable is in the form of "continuous data", so -> Regression analysis proceeds
  -> The predicted value of the regression analysis will be the "estimated value" (not the exact value, but there is an error)
"""
```

```python
### Visualize the length and weight of a perch in a scatterplot graph
# - interpret it

plt.title("Length and Weight of Perch")
plt.scatter(perch_length, perch_weight, c="red")

plt.xlabel("Length")
plt.ylabel("Weight")
plt.grid(linestyle="--")
plt.show

### <Analysis>
# - This data shows that as the length (x-axis) increases, the weight (y-axis) also increases.
# - It shows a linear distribution in the positive direction on the right-hand side.
# - When examining the relationship between length and weight, there is a positive correlation. 
#  (A positive correlation shows an upward trend, while a negative correlation shows a downward trend.)
# - The shape of the scatterplot is linear and has a curved shape
# - If the equation of the linear is used among regression methods to predict weight using length
#   It is believed that it is possible to estimate the weight to be predicted

### Linear equations have slopes and intercepts.
# - Equation of the straight line (y) = ax + b
# - Equation of curve (y) = ax^2 + bx + c
# - The value we know from the linear equation is : x value
# - The values that the model finds on its own are: slope, intercept

### The regression model self-judges the shape of the linearity and finds the slope and intercept on its own.
# -> Judging and finding for yourself is called training (model)
```

### Separating Data

```python
### Separate to training and testing data
# - Separation ratio : 75% : 25%
# - Random rule: 42
# - Regression does not require bias to be taken into account.
# - Variable name to use : train_input, train_target, test_input, test_target
train_input, test_input, train_target, test_target= train_test_split(
         perch_length, perch_weight, test_size=0.25, random_state=42
)

print(train_input.shape, train_target.shape)
print(test_input.shape, test_target.shape)
```

```python
### Training and testing independent variables are one dimension
# - In machine learning, the dimension of the independent variable must be two dimensional.
# - To make the dimensions of training and testing independent variables two dimensional...

### Using the Numfi function to change to two dimensions: reshape
# First value: -1 means the entire row
# Second value: means the column to be created
# Reshape(-1, 1): train_input Create 1 column for the entire row of data (create row, column)

# Converting training data to a 2 dimensional array
train_input = train_input.reshape(-1, 1)  
test_input = test_input.reshape(-1, 1)   
```

```python
print(train_input.shape, test_input.shape)
```

```python
### The final original
print(train_input.shape, train_target.shape)
print(test_input.shape, test_target.shape)
```

### Training with a regression model

```python
### Create Training Model
knr = KNeighborsRegressor()

### Training the model
knr.fit(train_input, train_target)

### Check coefficient of determination(r2) by Training Data and Testing Data
train_score = knr.score(train_input, train_target)
test_score = knr.score(test_input, test_target)

### Make predictions using the test data
test_pred = knr.predict(test_input)

print(f"training coefficient of determination: {train_score}, \nTest coefficient of determination: {test_score}, \n{test_pred} \n{test_target}")
```

```python
"""
<The coefficient of determination (accuracy) of regression models>
 - The regression model does not refer to it as accuracy, but is referred to as "determinism".
 - Another expression of coefficient of determination: it's also called determination, explanatory power, persuasive power.
 - Range of values for coefficients of determination: values between 0 and 1 (the closer to 1, the better)
 - When writing the result document: Write that the coefficient of determination of training is 0.97, 
   which is considered a very explanatory model.
 - Used as a criterion for determining whether the model is good or bad.
 - It is expressed as a model with an error of +- 0.03.
"""
```

### Evaluate model performance using accuracy : Determination of overfitting

```python
"""
<Common applications for both regression and classification analysis>
 - Comparison of scores using training data and scores using verification data after training(fit)
 - If there are many differences based on the comparison results --> "Overfitting occurs".
 - Overfitting terms: overfitting, underfitting, generalizing
 - Overfitting criteria: There is no exact answer (usually 0.1 difference will lead to overfitting)

 - Overfitting Classification
   - Overfitting
     * Training Accuracy > Validation (or Test) Accuracy
       -> If training accuracy is 1: Model not available
       -> If there is a difference of 0.1 or more between training accuracy and verification (or test) accuracy
       -> How to handle overfitting
          1. Increase the amount of data (requires additional collection)
             -> There's difficulty in data collection.
          2. Hyperparameter tuning
             -> Generally, start by processing in advance.
          3. Find another model
             -> If it can't be handled with 1 or 2... Use a different model

   - Underfitting
     * Training Accuracy < Validation (or Test) Accuracy
       -> Model not available
       -> How to handle underfitting
          1. Increase the amount of data (requires additional collection)
             -> There's difficulty in data collection.
          2. Hyperparameter tuning
             -> Generally, start by processing in advance.
          3. Find another model
             -> If it can't be handled with 1 or 2... Use a different model

   - Generalizing
     -> If the difference between training accuracy and verification (or test) accuracy is within 0.01~0.09

 - Terms expressed in a document
   - Assuming that it is generalized, the following terms are mainly used.
     * Training Accuracy(score) above 0.95: Very good model
     * Training Accuracy(score) above 0.90: Great model
     * Training Accuracy(score) 0.85 or higher: Good model
     * Training Accuracy(score) less than 0.85: Do not use the expression "good"

 - Most ideal accuracy (score)
   - Training Accuracy > Verification Accuracy > Test Accuracy (Most Ideal)
   - Training Accuracy > Verification Accuracy < Test Accuracy (Approved when it comes out like this)

 - tasks to create the most ideal accuracy
   - Pre-processing: Deflection processing, proper data separation, scaling
   - Hyperparameter tuning: modify parameter values that affect the performance of each model
   - Data Collection: Quantitative Collection
"""
```

```python
print(f"training coefficient of determination: {train_score}, \nTest coefficient of determination: {test_score}")
```

```python
### Training score is smaller than test score -> Underfitting occurs
# - KNN regression model cannot be used to predict weight using the length of perch as it is now
```

```python
"""
<Handling Underfitting>
 1. Increase the amount of data (requires additional collection)
  -> There's difficulty in data collection. -> x
 2. Hyperparameter tuning -> Can Be Advanced from the Current State
  -> Generally, start by processing in advance.
 3. Find another model -> Future Steps
  -> If it can't be handled with 1 or 2... Use a different model
"""
```

### Tuning Hyperparameter

```python
### Each model has different parameters to tune.
# - Parameters to tune by default in KNN: Number of Neighbor
"""
<How to Tuning>
 1. Create Model
 2. Training for the first time

 3. Changing the hyperparameter value
    -> Training model
 4. Check accuracy
 5. Repeat 3 to 4 times
"""
```

```python
### Retrain after changing the value of the hyperparameter using the existing trained model as is
# - Refrain after defining the number of Piper parameters as 5 neighbors
knr.n_neighbors = 5

# - Training the model
knr.fit(train_input, train_target)

# - Check training accuracy and test accuracy
train_score = knr.score(train_input, train_target)
test_score = knr.score(test_input, test_target)

# - Check training vs test difference
print(f"Training: {train_score}, Test: {test_score}, Training-Test: {train_score-test_score}")
```

```python
### Retrain after changing the value of the hyperparameter using the existing trained model as is
# - Refrain after defining the number of Piper parameters as 5 neighbors
knr.n_neighbors = 3

# - Training the model
knr.fit(train_input, train_target)

# - Check training accuracy and test accuracy
train_score = knr.score(train_input, train_target)
test_score = knr.score(test_input, test_target)

# - Check training vs test difference
print(f"Training: {train_score}, Test: {test_score}, Training-Test: {train_score-test_score}")
```

```python
### Find the number of neighbors that can show better vocabulary than when the number of neighbors is 3
### Lists to store training and test scores

# - List variable for the number of neighbors to use on the x-axis
neighbor = []
# - List variables to contain training and test accuracy for the y-axis
train_scores = []
test_scores = []

# Loop over the different values of k (neighbors)
for n in range(3, len(train_input) + 1, 2):
    knr.n_neighbors = n

    # - Training the model
    knr.fit(train_input, train_target)

    # - Check training accuracy and test accuracy
    train_score = knr.score(train_input, train_target)
    test_score = knr.score(test_input, test_target)

    # - Append the scores to the lists
    neighbor.append(n)
    train_scores.append(train_score)
    test_scores.append(test_score)

    # Print the results
    print(f"[{n}], Training: {train_score}, Test: {test_score}, Training-Test: {train_score - test_score}")
```

```python
### Create a list to show the range of values on the x-axis in detail
x_index = np.arange(3, len(train_input)+1, 2)
x_index
```

```python
### The results of training the number of 3 to the number of training data in the neighborhood
# - Visually check the training accuracy and test accuracy of the sun graph
# - x-axis is the number of neighbors, y-axis is the exact number of training and testing
# - Compare the training accuracy line and the test accuracy line on one graph
plt.title('Training vs Test Accuracy for Different Number of Neighbors')

plt.plot(neighbor, train_scores, color="blue", label="Training Accuracy")
plt.plot(neighbor, test_scores, color="red", label="Test Accuracy")

### Show x-axis values in more detail
plt.xticks(x_index)

plt.xlabel('Number of Neighbors')
plt.ylabel('Accuracy')
plt.grid(linestyle = "--")

plt.legend()
plt.show()
```

```python
# Use numpy to generate the range of neighbors
neighbors = np.arange(3, len(train_input) + 1, 2)

# Initialize empty arrays for the scores
train_scores = np.zeros(len(neighbors))
test_scores = np.zeros(len(neighbors))

# Loop over the different values of k (neighbors)
for i, n in enumerate(neighbors):
    knr.n_neighbors = n

    # Train the model
    knr.fit(train_input, train_target)

    # Check training and test accuracy
    train_scores[i] = knr.score(train_input, train_target)
    test_scores[i] = knr.score(test_input, test_target)

    # Print the results
    print(f"[{n}], Training: {train_scores[i]}, Test: {test_scores[i]}, Training-Test: {train_scores[i] - test_scores[i]}")

# Plotting the results
plt.title('Training vs Test Accuracy for Different Number of Neighbors')
plt.plot(neighbors, train_scores, color="blue", label="Training Accuracy")
plt.plot(neighbors, test_scores, color="red", label="Test Accuracy")

plt.xlabel('Number of Neighbors')
plt.ylabel('Accuracy')
plt.grid(linestyle="--")
plt.legend()
plt.show()
```

```python
### (Interpretation)
# - When using lengths of 50cm, 100cm, and 1000cm, 
#   an issue arises where the predicted weight values are all the same.
# - The reason for this is:
#   - When making predictions with arbitrary length values outside the range of the training data's minimum and maximum lengths,
#   - Only the nearest neighbors are selected, causing the prediction to always rely on the closest ones. 
#     In other words, even if the distance is far, the model perceives it as close.
# - Conclusion:
#   - The KNN model cannot accurately predict values outside the range of the training data used during training 
#     (it can make a prediction, but it cannot be trusted for accuracy).
#   - This is a limitation of the KNN algorithm, and there is no solution to overcome this.
#   - To use KNN effectively, the training data must be restricted to a range within the minimum and maximum values.

# - Other models developed after KNN do not have this limitation and should be used to overcome this issue.
```
